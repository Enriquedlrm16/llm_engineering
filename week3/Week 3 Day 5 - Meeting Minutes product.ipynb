# STEP 2: Analyze & Report

```python
system_message = """
You produce minutes of meetings from transcripts, with summary, key discussion points,
takeaways and action items with owners, in markdown format without code blocks.
"""

user_prompt = f"""
Below is an extract transcript of a Denver council meeting.
Please write minutes in markdown without code blocks, including:
- a summary with attendees, location and date
- discussion points
- takeaways
- action items with owners

Transcription:
{transcription}
"""

messages = [
    {"role": "system", "content": system_message},
    {"role": "user", "content": user_prompt}
  ]
```

```python
quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_compute_dtype=torch.bfloat16,
    bnb_4bit_quant_type="nf4"
)
```

```python
tokenizer = AutoTokenizer.from_pretrained(LLAMA)
tokenizer.pad_token = tokenizer.eos_token
inputs = tokenizer.apply_chat_template(messages, return_tensors="pt").to("cuda")
streamer = TextStreamer(tokenizer)
model = AutoModelForCausalLM.from_pretrained(LLAMA, device_map="auto", quantization_config=quant_config)
outputs = model.generate(inputs, max_new_tokens=2000, streamer=streamer)
```

# Student contribution

Student Emad S. has made this powerful variation that uses `TextIteratorStreamer` to stream back results into a Gradio UI, and takes advantage of background threads for performance! I'm sharing it here if you'd like to take a look at some very interesting work. Thank you, Emad!

```python
"""
Solución mejorada para generación de actas de reuniones con streaming
Utiliza TextIteratorStreamer y Gradio para una experiencia interactiva en tiempo real
"""

import gradio as gr
from transformers import AutoTokenizer, AutoModelForCausalLM, TextIteratorStreamer
import torch
from threading import Thread
import warnings
warnings.filterwarnings('ignore')

# Configuración del modelo
MODEL_NAME = "meta-llama/Llama-3.2-1B-Instruct"

class MeetingMinutesGenerator:
    """
    Generador de actas de reuniones con capacidad de streaming
    """

    def __init__(self, model_name=MODEL_NAME):
        """
        Inicializa el modelo y tokenizer

        Args:
            model_name: Nombre del modelo de HuggingFace a utilizar
        """
        print("🚀 Cargando modelo y tokenizer...")
        self.device = "cuda" if torch.cuda.is_available() else "cpu"
        print(f"📱 Usando dispositivo: {self.device}")

        self.tokenizer = AutoTokenizer.from_pretrained(model_name)
        self.model = AutoModelForCausalLM.from_pretrained(
            model_name,
            torch_dtype=torch.float16 if self.device == "cuda" else torch.float32,
            device_map="auto"
        )

        # Configurar pad token
        if self.tokenizer.pad_token is None:
            self.tokenizer.pad_token = self.tokenizer.eos_token

        print("✅ Modelo cargado exitosamente!")

    def create_prompt(self, transcript: str, language: str = "english") -> str:
        """
        Crea el prompt para el modelo

        Args:
            transcript: Transcripción de la reunión
            language: Idioma para el formato (english/spanish)

        Returns:
            Prompt formateado
        """
        if language.lower() == "spanish":
            system_message = """Eres un asistente experto en crear actas de reuniones profesionales.
Tu tarea es analizar la transcripción proporcionada y crear un acta estructurada."""

            user_message = f"""Por favor, analiza la siguiente transcripción de reunión y crea un acta profesional
que incluya:

1. **Resumen Ejecutivo**: Breve descripción general
2. **Información de la Reunión**: Asistentes, ubicación, fecha, tema
3. **Puntos de Discusión**: Temas principales tratados
4. **Conclusiones Clave**: Decisiones importantes tomadas
5. **Acciones Pendientes**: Tareas asignadas con responsables

Transcripción:
{transcript}

Acta de Reunión:"""
        else:
            system_message = """You are an expert assistant at creating professional meeting minutes.
Your task is to analyze the provided transcript and create structured meeting minutes."""

            user_message = f"""Please analyze the following meeting transcript and create professional
meeting minutes that include:

1. **Executive Summary**: Brief overview
2. **Meeting Information**: Attendees, location, date, topic
3. **Discussion Points**: Main topics covered
4. **Key Takeaways**: Important decisions made
5. **Action Items**: Assigned tasks with responsible parties

Transcript:
{transcript}

Meeting Minutes:"""

        # Formato de chat para Llama
        messages = [
            {"role": "system", "content": system_message},
            {"role": "user", "content": user_message}
        ]

        return self.tokenizer.apply_chat_template(
            messages,
            tokenize=False,
            add_generation_prompt=True
        )

    def generate_streaming(
        self,
        transcript: str,
        language: str = "english",
        max_new_tokens: int = 1000,
        temperature: float = 0.7,
        top_p: float = 0.9
    ):
        """
        Genera actas con streaming en tiempo real

        Args:
            transcript: Transcripción de la reunión
            language: Idioma del formato
            max_new_tokens: Número máximo de tokens a generar
            temperature: Temperatura para la generación
            top_p: Parámetro top-p para nucleus sampling

        Yields:
            Tokens generados en tiempo real
        """
        # Crear prompt
        prompt = self.create_prompt(transcript, language)

        # Tokenizar
        inputs = self.tokenizer(
            prompt,
            return_tensors="pt",
            truncation=True,
            max_length=2048
        ).to(self.device)

        # Configurar streamer
        streamer = TextIteratorStreamer(
            self.tokenizer,
            skip_prompt=True,
            skip_special_tokens=True
        )

        # Configurar parámetros de generación
        generation_kwargs = {
            **inputs,
            "streamer": streamer,
            "max_new_tokens": max_new_tokens,
            "temperature": temperature,
            "top_p": top_p,
            "do_sample": True,
            "pad_token_id": self.tokenizer.pad_token_id,
            "eos_token_id": self.tokenizer.eos_token_id,
        }

        # Iniciar generación en thread separado
        thread = Thread(target=self.model.generate, kwargs=generation_kwargs)
        thread.start()

        # Yield tokens a medida que se generan
        generated_text = ""
        for new_text in streamer:
            generated_text += new_text
            yield generated_text

        thread.join()


def create_gradio_interface():
    """
    Crea la interfaz de Gradio para la aplicación
    """
    # Inicializar generador
    generator = MeetingMinutesGenerator()

    def process_transcript(transcript, language, max_tokens, temperature, top_p):
        """
        Procesa la transcripción y genera actas
        """
        if not transcript.strip():
            yield "⚠️ Por favor, proporciona una transcripción válida."
            return

        yield "🔄 Generando acta de reunión...\n\n"

        for output in generator.generate_streaming(
            transcript=transcript,
            language=language,
            max_new_tokens=int(max_tokens),
            temperature=temperature,
            top_p=top_p
        ):
            yield output

    # Ejemplos de transcripciones
    example_spanish = """Reunión del Consejo de Denver
Asistentes: Miembros del Consejo
Ubicación: Ayuntamiento de Denver
Fecha: 27 de enero de 2024
Tema: Discusión sobre el logo de la Semana de la Confluencia

El logo representa la confluencia de dos ríos importantes. El agua es un elemento central
del diseño, simbolizando la importancia crítica del agua en la política global actual.
La parte posterior del logo está diseñada para representar el agua, reflejando las
situaciones actuales donde el agua es un tema crucial."""

    example_english = """Denver Council Meeting
Attendees: Council Members
Location: Denver City Hall
Date: January 27, 2024
Topic: Discussion on Confluence Week Logo

The logo represents the confluence of two major rivers. Water is a central element
of the design, symbolizing the critical importance of water in current global politics.
The back of the logo is designed to represent water, reflecting current situations
where water is a crucial issue."""

    # Crear interfaz
    with gr.Blocks(title="Meeting Minutes Generator", theme=gr.themes.Soft()) as demo:
        gr.Markdown("""
        # 📝 Generador de Actas de Reuniones con Streaming

        Esta aplicación utiliza **TextIteratorStreamer** y procesamiento en background threads
        para generar actas profesionales de reuniones en tiempo real.

        ### Características:
        - ⚡ Generación en tiempo real con streaming
        - 🌍 Soporte para inglés y español
        - 🎛️ Parámetros ajustables de generación
        - 🧵 Procesamiento eficiente con threads
        """)

        with gr.Row():
            with gr.Column(scale=1):
                transcript_input = gr.Textbox(
                    label="📄 Transcripción de la Reunión",
                    placeholder="Pega aquí la transcripción de tu reunión...",
                    lines=15,
                    max_lines=20
                )

                with gr.Accordion("⚙️ Configuración Avanzada", open=False):
                    language_select = gr.Radio(
                        choices=["english", "spanish"],
                        value="english",
                        label="🌐 Idioma del formato"
                    )

                    max_tokens_slider = gr.Slider(
                        minimum=200,
                        maximum=2000,
                        value=1000,
                        step=100,
                        label="📏 Tokens máximos"
                    )

                    temperature_slider = gr.Slider(
                        minimum=0.1,
                        maximum=1.5,
                        value=0.7,
                        step=0.1,
                        label="🌡️ Temperatura (creatividad)"
                    )

                    top_p_slider = gr.Slider(
                        minimum=0.1,
                        maximum=1.0,
                        value=0.9,
                        step=0.05,
                        label="🎯 Top-p (diversidad)"
                    )

                generate_btn = gr.Button("🚀 Generar Acta", variant="primary", size="lg")

                gr.Markdown("### 📋 Ejemplos:")
                gr.Examples(
                    examples=[
                        [example_spanish, "spanish"],
                        [example_english, "english"]
                    ],
                    inputs=[transcript_input, language_select],
                    label="Clic para cargar un ejemplo"
                )

            with gr.Column(scale=1):
                output_text = gr.Textbox(
                    label="📋 Acta Generada",
                    lines=25,
                    max_lines=30,
                    interactive=False
                )

                gr.Markdown("""
                ### 💡 Consejos:
                - La generación usa streaming, verás el texto aparecer en tiempo real
                - Ajusta la temperatura para controlar la creatividad
                - Un top-p más bajo genera texto más enfocado
                - Los tokens máximos controlan la longitud del documento
                """)

        # Conectar eventos
        generate_btn.click(
            fn=process_transcript,
            inputs=[
                transcript_input,
                language_select,
                max_tokens_slider,
                temperature_slider,
                top_p_slider
            ],
            outputs=output_text
        )

        gr.Markdown("""
        ---
        ### 📚 Información Técnica:

        **Implementación:**
        - Modelo: Llama-3.2-1B-Instruct
        - Streaming: TextIteratorStreamer
        - Procesamiento: Background threads
        - Framework: Gradio + Transformers

        **Autor de la variación original:** Emad S.
        """)

    return demo


def main():
    """
    Función principal para ejecutar la aplicación
    """
    print("=" * 60)
    print("📝 GENERADOR DE ACTAS DE REUNIONES CON STREAMING")
    print("=" * 60)

    # Crear y lanzar interfaz
    demo = create_gradio_interface()

    print("\n✅ Interfaz lista. Iniciando servidor Gradio...")
    print("🌐 Accede a la aplicación en el enlace que aparecerá abajo.\n")

    demo.launch(
        share=True,  # Crear link público
        server_name="0.0.0.0",
        server_port=7860,
        show_error=True
    )


if __name__ == "__main__":
    main()
```
